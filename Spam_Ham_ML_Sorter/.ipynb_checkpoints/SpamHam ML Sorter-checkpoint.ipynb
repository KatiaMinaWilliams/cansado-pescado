## Some Helpful Functions and Sanity Checks:

Listed together for brevity

```python
def words_in_texts(words, texts):
    '''
    Args:
        words (list-like): words to find
        texts (Series): strings to search in
    
    Returns:
        NumPy array of 0s and 1s with shape (n, p) where n is the
        number of texts and p is the number of words.
    '''
    indicator_array = np.array([[int(word in text) for word in words] for text in texts])
    return indicator_array

# If this doesn't error, your function outputs the correct output for this example
assert np.allclose(words_in_texts(['hello', 'bye', 'world'], 
                                  pd.Series(['hello', 'hello world hello'])),
                   np.array([[1, 0, 0], 
                             [1, 0, 1]]))

def accuracy_cv(Xtrain, ytrain, cv=5, reg='l2'):
    preds = []
    for i in range(cv):
        X_train, X_test, y_train, y_test = train_test_split(Xtrain, ytrain, test_size=0.2)
        cv_model = lm.LogisticRegression(reg).fit(X_train, y_train)
        preds.append(sum(cv_model.predict(X_test) == y_test)/len(y_test))
    return np.mean(preds)

def accuracy_test(Xtrain, ytrain, Xtest, ytest, reg='l2'):
    model = lm.LogisticRegression(reg).fit(Xtrain, ytrain)
    return sum(model.predict(Xtest) == ytest)/len(ytest)

def add_col(Phi, col):
    return np.array([np.append(Phi[i],col[i]) for i in range(Phi.shape[0])])
```

```python
def words_to_percents(df, words, searchseries):
    #Array of data frame, array of words, array/series to search for words in
    #Returns df with columns : word, spam, prop
    
    nspam = len(df[df['spam'] == 1].index)
    nham = len(df[df['spam'] == 0].index)
    
    spammy = pd.DataFrame(words_in_texts(words, searchseries))
    spammy.columns = words
    spammy.reset_index(inplace=True)
    spammy['spam'] = df['spam'].reset_index().iloc[:,1]

    for word in words:
        spammy = spammy.replace({word:{1:word}})

    propor_spam = pd.DataFrame()    

    for word in words:
        with_word = spammy.loc[:,[word,'spam']]
        new_rows = with_word.drop(with_word[with_word[word] == 0].index)
        new_rows.columns = np.array(['word', 'spam'])
        new_rows['prop'] = new_rows['spam']/nspam
        new_rows = new_rows.replace({'prop':{0:1/nham}})
        propor_spam = propor_spam.append(new_rows, ignore_index=True)


    propor_spam = propor_spam.groupby(['word', 'spam']).sum()

    propor_spam.reset_index(inplace=True)
    propor_spam = propor_spam.replace({'spam':{0:'Ham', 1:'Spam'}}) 
    return propor_spam

prop_spam = words_to_percents(train, spammy_words, train.loc[:, 'email'])

sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.barplot(x = 'word', y = 'prop', hue='spam', data = prop_spam.iloc[:40,:])
ax.set(ylabel='Proportion of Emails', xlabel='', ylim=(0,1));
#I already thought this from the previous, but picking particular words looks like a good separating
```

```python
#Spam sometimes has long strings of numbers. Lets look at those counts
import re

#At least five digits
def count_regex(df, reg, colname):
    #Counts instances regex in each string entry of colname in df
    #str colname, str regex, df
    #Returns col with spam, and count of regex
    
    long_nums = re.compile(reg)
    instances = np.array([len(long_nums.findall(email)) for email in df[colname]])
    reg_df = pd.DataFrame(instances, df['spam'])
    reg_df = reg_df.reset_index()
    reg_df.columns = ['spam', 'count_reg']
    return reg_df
five_df = count_regex(train, '[0-9]{5}', 'email')
sns.distplot(five_df[five_df['spam'] == 1]['count_reg'], label='Spam')
sns.distplot(five_df[five_df['spam'] == 0]['count_reg'], label='Ham')
plt.xlim(0,100)
plt.legend();

#Eh...the beginning's goodish. Let's check some larger numbers before deciding which to use
```

```python
seven_df = count_regex(train, '[0-9]{7}', 'email')
sns.distplot(seven_df[seven_df['spam'] == 1]['count_reg'], label='Spam', kde=True)
sns.distplot(seven_df[seven_df['spam'] == 0]['count_reg'], label='Ham', kde=True)
plt.xlim(0,50)
plt.legend();
```

```python
#Trying: Instead of counts, just if they show up.
def has_regex(df, reg, colname):
    #Counts instances regex in each string entry of colname in df
    #str colname, str regex, df
    #Returns df with spam, regex, and prop of either spam or ham with that regex
    nspam = len(df[df['spam'] == 1].index)
    nham = len(df[df['spam'] == 0].index)
    
    long_nums = re.compile(reg)
    instances = np.array([int((len(long_nums.findall(email)) != 0)) for email in df[colname]])
    reg_df = pd.DataFrame(instances, df['spam'])
    reg_df = reg_df.reset_index()
    reg_df.columns = ['spam', 'has_reg']
    
    is_spam = reg_df[reg_df['spam'] == 1].copy()
    is_ham = reg_df[reg_df['spam'] == 0].copy()
    
    is_spam['has_reg'] = is_spam['has_reg']/nspam
    is_ham['has_reg'] = is_ham['has_reg']/nham
    reg_df = is_spam.append(is_ham).sort_index()
    reg_df = reg_df.groupby('spam').sum().reset_index()
    return reg_df


has_five = has_regex(train, '[0-9]{5}', 'email')
sns.barplot(x='spam', y='has_reg', data=has_five)
plt.ylim(0,1);
#It may get better with bigger numbers
```

```python
has_ten = has_regex(train, '[0-9]{10}', 'email')
sns.barplot(x='spam', y='has_reg', data=has_ten)
plt.ylim(0,1);
#Nope, it's actually worse

#Conclusion: make a "many big numbers" binary predictor that checks if an email has at least 20 numbers  each at least
#7 digits long
#This will most likely correspond to emails with lots of font color changes as well (since those color codes are six)
```

```python
#Best words
diffs_df = diffs_df[diffs_df['word'] != 'html']
best_words = diffs_df[diffs_df['Diff in Prop'] > .15]['word']
best_words
```

```python
### Symbols in Subject ###

symbols = np.array(['%','$', '!','&','@', '#', '*', '(', ')', '=','+', '<', '{'])

spammy_subs = pd.DataFrame(words_in_texts(symbols, subjects))
spammy_subs.columns = symbols
spammy_subs.reset_index(inplace=True)
spammy_subs['spam'] = train['spam'].reset_index().iloc[:,1]

for word in symbols:
    spammy_subs = spammy_subs.replace({word:{1:word}})

prop_spam = pd.DataFrame()    

for word in symbols:
    with_word = spammy_subs.loc[:,[word,'spam']]
    new_rows = with_word.drop(with_word[with_word[word] == 0].index)
    new_rows.columns = np.array(['word', 'spam'])
    new_rows['prop'] = new_rows['spam']/nspam
    new_rows = new_rows.replace({'prop':{0:1/nham}})
    prop_spam = prop_spam.append(new_rows, ignore_index=True)


prop_spam = prop_spam.groupby(['word', 'spam']).sum()

prop_spam.reset_index(inplace=True)
prop_spam = prop_spam.replace({'spam':{0:'Ham', 1:'Spam'}})

diffs = [abs(prop_spam['prop'][i] - prop_spam['prop'][i+1]) for i in np.arange(0, len(prop_spam.iloc[:,0]), 2)] 
diffs_df = pd.DataFrame(prop_spam['word'].unique(), diffs).reset_index().sort_values(by='index', ascending=False)
diffs_df.columns = ['Diff in Prop', 'word']
diffs_df[:20]
```
